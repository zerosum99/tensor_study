```python
import tensorflow as tf
```

```python
sess = tf.InteractiveSession()
```

# 1.  변수 저장하기

```python
raw_data = [1.,2.,8,-1.,0.,5.5, 6.,13]
```

```python
help(tf.Variable)
```

### 리스트에 곱셈 연산은 기존 리스트의 원소를 곱셈 다음이 숫자값만큼 만든다

```python
len(raw_data)
```

```python
[False] * len(raw_data)
```

  ## 변수 생성하기

```python
spikes = tf.Variable([False] * len(raw_data), name='spikes',  use_resource=True)
```

### 변수는 초기화 없이 실행하면 예외 발생

```python
#. spikes.eval()
```

### 변수를 초기화 : 메소드 run을 처리

```python
dir(spikes.initializer)
```

```python
spikes.initializer.run()
```

```python
spikes.eval()
```

### 변수 저장 및 복구 가능할 객체 생성

```python
saver = tf.train.Saver()
```

```python
for i in range(1, len(raw_data) ) :
    
    if raw_data[i]  - raw_data[i-1] > 5 :
        spikes_val = spikes.eval()
        
        spikes_val[i] = True
        updater = tf.assign(spikes, spikes_val)
        updater.eval()
        
```

###  실행된 세션 내의 변수 등을 저장하기

```python
save_path = saver.save(sess, "./saved/train2")
print(" spikes data saved in file : %s" % save_path)

sess.close()
```

### 저장된 디렉토리 내의 파일 확인하기

```python
%ls ./saved
```

# 2. 변수 불러오기

## 다른 세션을 만든다

```python
sess2 = tf.InteractiveSession()
```

## 다시 네트워크 생성

- meta파일을 생성 했으므로 이것을 불러오는 방식으로 network을 재생성 할 수 있다.
- meta파일을 불러오기 위해서는 tf.train.import_meta_graph() 함수를 이용한다.


## 저장된 메타 파일에서 그래프를 다시 만든다.

```python
saver = tf.train.import_meta_graph('./saved/train2.meta')
```

```python
saver
```

```python
 sess2.run(tf.global_variables_initializer())
```

## 저장된 체크포인트 파일 확인하기

```python
ckpt_state = tf.train.get_checkpoint_state("saved")

print(type(ckpt_state))
```

```python
print("첫번째 정보 사용법:", ckpt_state.model_checkpoint_path)
print("두번째 정보 사용법:", ckpt_state.all_model_checkpoint_paths)
```

## 저장된 정보 재적재하기

```python
help(saver.restore)
```

```python
saver.restore(sess2, tf.train.latest_checkpoint("saved"))
```

```python
saver.restore(sess2,  "./saved/train2")
```
